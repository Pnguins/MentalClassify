```python
import uuid
import numpy as np
from qdrant_client import QdrantClient
from itertools import islice

# Placeholder for the tokenizer function
def generate_embeddings(texts, tokenizer):
    """
    Generates embeddings for a list of texts using a tokenizer.

    Args:
        texts (list of str): List of input texts.
        tokenizer: Tokenizer object that supports embedding generation.

    Returns:
        list of list of float: List of embeddings.
    """
    embeddings = []
    for text in texts:
        # Generate token IDs for the text and flatten into a single list
        token_ids = tokenizer(text, return_tensors="pt", truncation=True, padding=True,max_length=1024)["input_ids"]
        embeddings.append(token_ids.numpy().flatten().tolist())
    return embeddings

def pad_embeddings(embeddings, target_size):
    """
    Pads each embedding to the target size with zeros.

    Args:
        embeddings (list of list of float): List of embeddings (vectors).
        target_size (int): The required dimensionality.

    Returns:
        list of list of float: List of padded embeddings.
    """
    padded_embeddings = [
        np.pad(embedding, (0, max(0, target_size - len(embedding))), mode='constant').tolist()
        for embedding in embeddings
    ]
    return padded_embeddings

VECTOR_SIZE = 1024  # Target vector size for Qdrant collection

# Generate embeddings and pad them to the required size
print("Generating and padding embeddings...")
df['embedding'] = generate_embeddings(df['post'].tolist(), tokenizer)
df['embedding'] = pad_embeddings(df['embedding'], VECTOR_SIZE)

# Prepare points for Qdrant insertion
print("Preparing points for Qdrant...")
points = [
    {
        "id": str(uuid.uuid4()),  # Unique ID for each point
        "vector": embedding,      # Embedding vector
        "payload": {
            "post": row['post'],   # Include text
            "Label": row['Label'],  # Include label
            "Reasoning": row["Reasoning"],
            "DS": row["DS"]
        }
    }
    for embedding, (_, row) in zip(df['embedding'], df.iterrows())
]

# Define batch size (adjust as needed to keep under 33 MB limit)
BATCH_SIZE = 100

# Helper function to split points into batches
def batch_points(points, batch_size):
    """Yield successive n-sized chunks from points."""
    it = iter(points)
    for batch in iter(lambda: list(islice(it, batch_size)), []):
        yield batch

# Upsert into Qdrant in batches
collection_name = "Mental_Llama"
print(f"Upserting points into Qdrant collection: {collection_name} in batches...")
for batch in batch_points(points, BATCH_SIZE):
    qdrant_client.upsert(collection_name=collection_name, points=batch)
print("Points successfully inserted!")
```
